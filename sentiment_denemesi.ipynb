```python
from transformers import pipeline
```

```python
pipe = pipeline("text-classification", model="savasy/bert-base-turkish-sentiment-cased")
```

```python
yorum = "Fiber 100mb SuperOnline kullanıcısıyım yaklaşık 2 haftadır @Twitch @Kick_Turkey gibi canlı yayın platformlarında 360p yayın izlerken donmalar yaşıyoruz.  Başka hiç bir operatörler bu sorunu yaşamazken ben parasını verip alamadığım hizmeti neden ödeyeyim ? @Turkcell "
```

```python
pipe(yorum)
```
